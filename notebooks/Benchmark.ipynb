```python
import time
import sarge
import seaborn as sns
import shutil
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
os.chdir('..')
```

```python
'''
Wrapper for timing fastq-dump under default settings.
'''
def fastq_dump_timed(sra,scratch):
    
    cmd = 'fastq-dump {sra_id}'.format(sra_id=sra)
    print(cmd)
    
    start_time = time.time()
    sarge.run(cmd)
    finish_time = time.time()-start_time
    os.remove(sra+'.fastq')
    shutil.rmtree(scratch)
    
    return finish_time
```

```python
'''
Wrapper for timing the modified fastq-dump with variable number of cores.
'''
def modified_fastq_dump_timed(sra,scratch,num_cores):
    
    cmd = 'modified_fastq_dump.py -s {sra_id} -n {cpu} --scratch {scratch_dir}'.format(sra_id=sra,cpu=num_cores,
                                                                                       scratch_dir=scratch)
    print(cmd)
    start_time = time.time()
    sarge.run(cmd)
    finish_time = time.time()-start_time
    os.remove(sra+'.fastq')
    
    return finish_time
```

```python
all_times = []
```

```python
for num in [1,2,5,10,20]:
    for i in range(5): 
        %time all_times.append({'Time':modified_fastq_dump_timed('SRR836087','/gpfs/data01/sjroth/temp_scratch',num),'Number of Cores':num})
```

```python
all_times = pd.DataFrame(all_times)
all_times
```

```python
sns.set()
sns.lmplot(x='Number of Cores',y='Time',data=all_times,x_estimator=np.mean,order=2)
```
